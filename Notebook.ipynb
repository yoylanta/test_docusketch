# MAIN IDEA
#### The main idea of my visualization analysis is to examine the differences beetween corresponding CEILING and FLOOR deviation values. I assume that the outputs of these corresponding variables, e.g. "ceiling_max" and "floor_max" or "ceiling_min" and "floor_min", ideally should have been equal to each other. 
#### Additionally, bringing to attention the fact that all deviation values in the original data set are positive, it is fair to assume that they are actually absolute values of deviations.
#### For the sake of visual interpretability, the histogram chart choice is justified, it is a common practice to use it when determining whether the outputs of two data are different or not is required.


# VISUALIZATION ANALYSIS

### **Mean deviation distribution plot**

```python
plot_creator.show_hist("mean", logy=True)
```

#### MAD distribution indicates the average differenece between observations. Our data is pretty clustered around the mean and is not spread out.

## **Min, max and mean deviation differences**

### **Min deviations difference**

```python
plot_creator.show_hist("min_diff", logy=True)
```

#### As you can see, there is a little asymmetry in distribution, that could possibly mean that there is some inconsistency in predictions that model performs depending on whether it is floor or ceiling predictions.

### **Max deviations difference**

```python
plot_creator.show_hist("max_diff", logy=True)
```

#### Yet another slight asymmetry

### **Mean deviations difference**

```python
plot_creator.show_hist("mean_diff", logy=True)
```

#### Once again, we observe a little bit of asymmetry in distribution, which possibly means that the model slightly makes a systematic error in its predictions in relation to whether it ceiling or floor predictions.

#### Now let's visualize the distributions of floor and ceiling means separately

```python
plot_creator.show_hist("ceiling_mean", logy=True, show_std=False)
```

```python
plot_creator.show_hist("floor_mean", logy=True, show_std=False)
```

```python
plot_creator.show_hist("ceiling_mean", logy=True, show_std=False), plot_creator.show_hist("floor_mean", logy=True, show_std=False)
```
